```python
# Merge data of CSV files we are going to use
import pandas as pd

# Define data path
dataPath = '../../Data/Data_Lloc_naix_regio/'
# Define file paths for each year
def merge_csvs_explicit():
    years = range(1997, 2024)  # Only up to 2023 (existing data)
    
    dfs = []
    for year in years:
        file_path = f"{dataPath}{year}_pad_mdb_lloc-naix-regio_sexe.csv"
        try:
            df = pd.read_csv(file_path, delimiter=',')
            # Replace '..' with 0 in column 'Valor' and coerce to numeric
            if 'Valor' in df.columns:
                df['Valor'] = df['Valor'].replace('..', 0)
                df['Valor'] = pd.to_numeric(df['Valor'], errors='coerce').fillna(0)
            dfs.append(df)
            print(f"Successfully read {year}")
        except FileNotFoundError:
            print(f"Warning: File for year {year} not found")
    
    if dfs:
        df_all_years = pd.concat(dfs, ignore_index=True)
        print(f"Successfully merged {len(dfs)} files")
        return df_all_years
    else:
        print("No files were found to merge")
        return None


# Concatenate all DataFrames into a single DataFrame
df_combined = merge_csvs_explicit()  # or merge_csvs_explicit()
# Save the combined DataFrame to a new CSV file
df_combined.to_csv(f"{dataPath}merged_lloc-naix-regio_sexe.csv", index=False)
print("Merged data saved to 'merged_lloc-naix-regio_sexe.csv'")

```
